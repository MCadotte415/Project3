```python
import csv
import sqlite3
```

```python
def my_decoder(s):
    return s.decode('utf-8')
```

```python
conn = sqlite3.connect('mydatabase.db')
conn.text_factory = my_decoder
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('''DROP TABLE IF EXISTS treat''')
```

```python

cursor.execute('''CREATE TABLE treat
                  (id integer Primary Key,
                  url text,
                  species text,
                  breed_primary text,
                  breed_mixed text,
                  color_primary text,
                  age text, 
                  sex text,
                  size text,
                  coat text,
                  fixed text,
                  house_trained text,
                  shots_current text,
                  name text,
                  contact_city text,
                  contact_state text,
                  contact_zip integer, 
                  description text)''')
```

```python
conn.commit()
```

```python
with open('state_dog.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        cursor.execute("INSERT INTO treat (id, url, species, breed_primary, breed_mixed, color_primary, age, sex, size, coat, fixed, house_trained, shots_current, name, contact_city, contact_state, contact_zip, description) VALUES (:id, :url, :species, :breed_primary, :breed_mixed, :color_primary, :age, :sex, :size, :coat, :fixed, :house_trained, :shots_current, :name, :contact_city, :contact_state, :contact_zip, :description)", row)
```

```python
conn.commit()
```

```python
#checking if the table was made
cursor.execute("SELECT name FROM treat")
rows = cursor.fetchall()


for row in rows:
    print(row)
```

```python
conn.close()
```
